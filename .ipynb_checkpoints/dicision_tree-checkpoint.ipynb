```python


used_features = []
time_counter = {'init':0, 'find':0, 'seperate':0, 'pre_cut':0, 'train':0, 'G_D_a':0, 'H_D':0, 'G_D_a_t':0}

def H_D(y):
    time_begin = time.time()
    r = 0
    l_y = len(y)
    log_2 = np.log(2)
    count = np.bincount(y)
    for i in set(y):
        r -= count[i]/l_y * np.log(count[i]/l_y) / log_2
    time_end = time.time()
    time_counter['H_D'] += time_end - time_begin
    return r

def T_a(y_a):
    l = [item for item in y_a]
    l.sort()
    l = [(l[i] + l[i + 1]) / 2 for i in range(len(l) - 1)]
    return l
    
def G_D_a_t(y,a,t,h_D):
    time_begin = time.time()
    pos_list, neg_list = [], []
    for i in range(len(a)):
        if a[i] > t:
            pos_list.append(y[i])
        else:
            neg_list.append(y[i])
    t = len(pos_list) / len(y) * H_D(pos_list)
    t += len(neg_list) / len(y) * H_D(neg_list)
    time_end = time.time()
    time_counter['G_D_a_t'] += time_end - time_begin
    return h_D - t

def G_D_a(y,a):
    time_begin = time.time()
    t_a = T_a(a)
    max_value = 0
    max_t = 0
    h_D = H_D(y)
    for t in t_a:
        v = G_D_a_t(y,a,t,h_D)
        if v > max_value:
            max_value = v
            max_t = t
    time_end = time.time()
    time_counter['G_D_a'] += time_end - time_begin
    return max_value, max_t
```

```python
import time

class Decision_tree():
    
    def __init__(self):
        pass
    
    def train(self, X, y):
        time_begin = time.time()
        rest_indexs = list(range(len(X[0])))
        self.dtn = Decision_tree_node(rest_indexs, X, y)
        self.dtn.train()
        time_end = time.time()
        print('Training takes',time_end - time_begin,'s')
        
    def predict(self, X):
        y = [self.dtn.predict(x) for x in X]
        return np.array(y)
        
        
class Decision_tree_node():
    
    def __init__(self, rest, X, y):
        time_begin = time.time()
        self.using_index, self.t = self.find_best_index(rest, X, y)
        self.rest = rest
        self.X, self.y = X, y
        self.pos = None
        #print('\tadd\t',used_features,self.using_index)
        used_features.append(self.using_index)
        time_end = time.time()
        time_counter['init'] += time_end - time_begin
    
    def find_best_index(self, rest, X, y):
        time_begin = time.time()
        max_index = -1
        max_value = -1
        max_t = 0
        for index in rest:
            if index in used_features:
                continue
            a = [item[index] for item in X]
            g, t = G_D_a(y,a) #
            if g > max_value:
                max_value = g
                max_index = index
                max_t = t
        time_end = time.time()
        time_counter['find'] += time_end - time_begin
        return max_index, max_t
    
    def seperate_data(self, X, y):
        time_begin = time.time()
        X_pos, X_neg, y_pos, y_neg = [], [], [], []
        index, t = self.using_index, self.t
        for i in range(len(X)):
            if X[i][index] > t:
                X_pos.append(X[i])
                y_pos.append(y[i])
            else:
                X_neg.append(X[i])
                y_neg.append(y[i])
        time_end = time.time()
        time_counter['seperate'] += time_end - time_begin
        return X_pos, X_neg, y_pos, y_neg
    
    def pre_cut(self, y_pos, y_neg):
        time_begin = time.time()
        count = {}
        for item in self.y:
            count.setdefault(item,0)
            count[item] += 1
        index_max = max(count, key=count.get)
        gener_per_cut = count[index_max] / len(self.y)
        count = {}
        for item in y_pos:
            count.setdefault(item,0)
            count[item] += 1
        index_pos_max = max(count, key=count.get)
        gener_per_not_cut = count[index_pos_max] / len(y_pos)
        for item in y_neg:
            count.setdefault(item,0)
            count[item] += 1
        index_neg_max = max(count, key=count.get)
        gener_per_not_cut += count[index_neg_max] / len(y_neg)
        self.index_max = index_max
        time_end = time.time()
        time_counter['pre_cut'] += time_end - time_begin
        return gener_per_cut < gener_per_not_cut
        # return false if need cut
    
    def train(self):
        time_begin = time.time()
        if H_D(self.y) == 0:
            self.pos = self.y[0]
        else:
            X_pos, X_neg, y_pos, y_neg = self.seperate_data(self.X, self.y)
            if self.pre_cut(y_pos, y_neg):
                #print(1)
                self.dtn_pos = Decision_tree_node(self.rest, X_pos, y_pos)
                self.dtn_pos.train()
                #print(2)
                self.dtn_neg = Decision_tree_node(self.rest, X_neg, y_neg)
                self.dtn_neg.train()
            else:
                self.pos = self.index_max
        #print('\tremove\t',used_features,self.using_index)
        time_end = time.time()
        used_features.remove(self.using_index)
        time_counter['train'] += time_end - time_begin
    
    def predict(self,x):
        #print(self.using_index)
        if self.pos != None:
            return self.pos
        elif x[self.using_index] > self.t:
            return self.dtn_pos.predict(x)
        else:
            return self.dtn_neg.predict(x)
            
        
```

```python
from sklearn.model_selection import KFold

class Mult_Decision_tree():
    
    def __init__(self, num = 5):
        self.num = num
    
    def train(self, X, y):
        kf = KFold(n_splits = self.num)
        self.dts = []
        for train, test in kf.split(X):
            X_kf = X[train]
            y_kf = y[train]
            dt = Decision_tree()
            dt.train(X_kf,y_kf)
            self.dts.append(dt)
        
    def predict(self,x):
        results = []
        result = []
        for dt in self.dts:
            results.append(dt.predict(x))
        for i in range(len(x)):
            value = {}
            for r in results:
                value.setdefault(r[i],0)
                value[r[i]] += 1
            value_max = max(value, key=value.get)
            result.append(value_max)
        return np.array(result)
            

```
